# Introduction

This notebook compares three algorithms for determining daylength. Two algorithms are undocumented: they came to me in source code form without a reference to their origin. The third algorithm is from Chapter 15 "Rising, Transit, and Setting", and Chapter 25 "Solar Coordinates" of [4]. The implementation of the third algorithm leverages the python AstroPy library [1], which in turn is built on the C library "Essential Routines for Fundamental Astronomy (ERFA)" [2], an open source version of the International Astronomical Union's official "Standards of Fundamental Astronomy (SOFA)" software library. [3]

Testing of the two undocumented algorithms proceeded by comparing the algorithm's daylength predictions to predictions from the US Naval Observatory's day length calculator [5]. Using Missoula, MT as a location predictions were compared for the first day of every other month.  These two algorithms agreed with the US Naval Observatory's predictions to within 20 minutes and 30 minutes, respectively. As these predictions were to be used to generate an index of photoperiod where the entire transition between "limiting" to "not limiting" occurs in a single hour, better agreement was deemed necessary.

Testing of the algorithm in [4] repeated the above, but with three locations: Missoula, MT, Prudhoe Bay, AK, and Tiksi, Siberia. The extra locations were chosen to check for proper behavior during those times of the year when the sun doesn't rise or set, and to check proper behavior on both sides of the prime meridian. Agreement throughout the year at Missoula, Prudhoe Bay, and Tiksi was 3.5, 9, and 3 minutes, respectively.

This notebook compares the algorithm of [4] to the undocumented algorithms over a wider range of latitudes and seasons. This is intended to give a better representation of the performance of the undocumented algorithms than a point comparison can provide.

# Basic Setup

```python
import os
os.chdir("..")
import sun
import astropy.units as u
import astropy.coordinates as c
import astropy.time as t
import met_functions as met
```

```python
latitudes = np.arange(-80, 81, 5) * u.deg
longitudes = np.zeros( latitudes.shape ) * u.deg
locations = c.EarthLocation.from_geodetic(longitudes, latitudes)
jan_1 = t.Time('2013-01-01')
doy = np.arange(0,365,10) * u.day
days = jan_1 + doy
shape = ( (doy.size, latitudes.size) ) 
daylength = u.Quantity(np.empty( shape ) , unit=u.hour)
correction = u.Quantity(np.empty(shape), unit=u.min)
```

```python
for i in range(len(days)) : 
    up = sun.Uptime(sun.SunPosition, days[i], locations) 
    daylength[i,:] = up.accurate_daylength()
    cor = up.correction() 
    effect = cor[:,2] + cor[:,1]
    correction[i,:] = effect
print np.abs(correction).max()
```

So at this point, we've built a "daylength" array using the algorithm in [4]. It has dimensions of (days, latitude). Longitude is excluded as not relevant. The algorithm in
[4] is iterative, first calculating an approximate solution, then calculating a correction, then calculating another correction if necessary. My implementation only performs the first correction. The figure of 19.434 minutes above represents the largest first order correction found on any day at any latitude. This indicates that the first order correction is important for the current application.

# Exploration of 1st order correction

One of the distinctions between the undocumented algorithms and the algorithm in [4] is the calculation of a correction to the initial approximation. Here we try to find out where the correction makes the most difference to the final answer.

```python
cs =contour(correction.to(u.min), extent=[-80,80,0,360],levels=np.arange(-15,15,2))
clabel(cs, fmt='%d')
```

In the above, it is clear that the first order correction makes the most difference near the poles near the season of "most sun". For our application (Russia), it looks like the bulk of the landmass has a difference of +/- 1 minute between initial approximation and final answer. This indicates that it may be possible to forego the overhead of calculating the "accurate" answer. The above contour plot also indicates that the maximum correction value of 19 minutes is abnormally large.

The next step is to locate that maximum correction value.

```python
i_max_cor = np.unravel_index(np.argmax(np.abs(correction)), correction.shape)
print i_max_cor
print doy[i_max_cor[0]]
print latitudes[i_max_cor[1]]
```

Day 320 at 70 deg south is the bright red diamond in the upper-left hand corner of the 
correction contour plot. 

To summarize this section, the magnitude of the correction factor is +/- 1 minute for the vast majority of latitudes throughout the year. The maximum correction value of 19 minutes is a highly localized phenominon, and occurs at a latitude where there are no plants to speak of. 

The improved agreement between the US Naval Observatory and the algorithm of [4] cannot be attributed to the correction factor. The bulk of the improvement must be due to a better initial approximation.

# Comparison with algorithm 1

The first undocumented algorithm came from Daymet (MT-CLIM). The python version has been tested using a test case included in the C version's comments. The python version behaves the same as the C version to the precision of the test case's answers. This section characterizes the magnitude of the difference between this algorithm and the algorithm of [4].

```python
mt_clim_daylength = u.Quantity(np.empty( shape ) , unit=u.hour)
for i in range(len(days)) :
    mt_clim_daylength[i,:] = met.calc_dayl(latitudes, doy[i])
    
diff = (daylength - mt_clim_daylength).to(u.min)
plot(doy, diff[:,4].to(u.min))
title("Daylength prediction difference vs. day of year (60 deg S)")
xlabel("Day of year")
ylabel("Difference in daylength (minutes)")
```

```python
maxdiff_lat = np.max(diff, axis=0)
plot(latitudes, maxdiff_lat)
title("Maximum daylength difference vs. latitude")
xlabel("Latitude (deg)")
ylabel("abs(MT-CLIM-AA) (minutes)")
```

```python
maxdiff_doy = np.max(diff, axis=1)
plot(doy, maxdiff_doy)
title("Maximum daylength difference vs. day of year")
xlabel("Day of year")
ylabel("abs(MT_CLIM-AA) (minutes)")
```

```python
figsize(8,6)
cs = contour(diff, extent=[-80,80,0,360], levels=np.arange(-10,60,3))
clabel(cs, fmt='%1.0f')
title("Daylight difference vs. latitude and day of year")
xlabel("Latitude (deg)")
ylabel("Day of year")
```

```python
mt_clim_clip = np.clip(mt_clim_daylength, 10*u.hour, 11*u.hour)
daylength_clip = np.clip(daylength, 10*u.hour, 11*u.hour)
mt_clim_diff_gsi = (mt_clim_clip - daylength_clip) / 1*u.hour
```

```python
figsize(8,6)
cs = contour(mt_clim_diff_gsi, extent=[-80,80,0,360], levels=np.arange(-1,1,0.1))
clabel(cs, fmt='%1.1f')
title("GSI index difference vs. latitude and day of year")
xlabel("Latitude (deg)")
ylabel("Day of year")
```

The previous two figures illustrate the difference between the MT-CLIM/Daymet algorithm and the algorithm of [4]. The difference in daylength is not easily specified as a single number. Error is larger in the northern hemisphere than in the southern hemisphere. To first order, the error surface is fairly flat around the equator, although the value of this flat error space is 5-8 minutes. In the northern hemishpere, error accumulates in a smooth gradient from the equator to the pole. Our region of interest (eastern Russia and Siberia) occupies a region of the planet where the error surface climbs rather steeply from 14 minutes to an hour. 

In terms of the GSI iPhoto index, however, only the span from 10-11 hours of daylight matters. The second plot shows that the difference in iPhoto caused by the difference between the two methods of calculating daylight primarily affects the spring and fall seasons. At low latitudes, the magnitude of error in iPhoto is 10-20% over a 1.5 month period, whereas in the northern latitudes error of 30-40% is concentrated into a 3 week period.

# Comparison with Algorithm 2

This section is essentially a repeat of the last section. The plots reveal some numerical instability with the implementation of algorithm 2 for the case when the sun is up all day or down all day. The error surface is substantially less flat around the equator, but the northern hemisphere appears qualitatively similar. 

The error surface for iPhoto has roughly the same structure and characteristics as the above. Autumn has a larger error further south than the comparison with algorithm 1.

```python
figsize(8,6)
cs = contour(alg2_diff, extent=[-80,80,0,360], levels=np.arange(-10,60,3))
clabel(cs, fmt='%1.0f')
title("Daylight difference vs. latitude and day of year")
xlabel("Latitude (deg)")
ylabel("Day of year")
```
